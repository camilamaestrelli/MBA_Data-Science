# **Projeção demográfica das Unidades da Federação do Brasil**

## **Entendendo o contexto**

### **Projeções Populacionais do IBGE 2010-2060**

Em 2018 o Instituto Brasileiro de Geografia e Estatística - IBGE, que realizou,projeções de populações para as unidades da federação do Brasil de 2010 a 2060.

Para tanto foi utilizado o método das componentes demográficas, onde se faz as projeções da população por faixas etárias quinquenais, baseando-se na Mortalidade, Migração e Fecundidade.

As tabelas com as populações projetadas, bem como os indicadores das componentes estão disponíveis de forma aberta no site do IBGE.

### **O Objetivo da Aula**

Construir uma visualização dinâmica e interativa de indicadores demográficos das Unidades da Federação do Brasil, de 2010 a 2060.

### **O dataset**

Baseado em informações do IBGE, este conjunto de dados foi estruturado de forma que cada instância representa uma Unidade da Federação do Brasil em determinado ano, e contém os seguintes atributos:

- **Obitos:** Número de óbitos

- **TBM:** Taxa Bruta de Mortalidade
- **EVN:** Expectativa de Vida ao Nascer
- **EVN_Homens:** Expectativa de Vida dos Homens
- **EVN_Mulheres:** Expectativa de Vida das Mulheres
- **TMI:** Taxa de Mortalidade Infantil
- **TMI_Homens:** Taxa de Mortalidade Infantil do sexo masculino
- **TMI_Mulheres:** Taxa de Mortalidade Infantil do sexo feminino
- **TFT:** Taxa de Fecundidade Total
- **TEF_15_19:** Taxa de Fecundidade Específica entre mulheres de 15 a 19 anos
- **TEF_20_24:** Taxa de Fecundidade Específica entre mulheres de 20 a 24 anos
- **TEF_25_29:** Taxa de Fecundidade Específica entre mulheres de 25 a 29 anos
- **TEF_30_34:** Taxa de Fecundidade Específica entre mulheres de 30 a 34 anos
- **TEF_35_39:** Taxa de Fecundidade Específica entre mulheres de 35 a 39 anos
- **TEF_40_44:** Taxa de Fecundidade Específica entre mulheres de 40 a 44 anos
- **TEF_45_49:** Taxa de Fecundidade Específica entre mulheres de 45 a 49 anos

- **Rz_Dep_Idosos:** Razão de Dependência dos Idosos
- **Rz_Dep_Total:** Razão de Dependência Total
- **Indice_Envelhecimento:** Índice de Envelhecimento
- **Proporcao_0a14anos:** Proporção da População de 0 a 14 anos
- **Proporcao_15a64anos:** Proporção da População de 15 a 64 anos

```python
# Importando o Pandas
import pandas as pd
```

```python
# Importando o conjunto de dados das projeções populacionais
projecao = pd.read_csv(
    filepath_or_buffer = '/content/projecao_IBGE_2018_atualizada06042020.csv', 
    sep=';',
    decimal=',',
    encoding = 'utf-8'
)
```

```python
# Espiando o conjunto de dados
projecao
```

### **Espiando a estrutura dos dados**

```python
# Informacoes gerais do conjunto de dados
projecao.info(verbose = True)
```

```python
# Verificando valores missing
projecao.isnull().sum()
```

### **Transformações no Conjunto de dados**

```python
# Filtrando somente as Unidades da Federação
projecao_UF = projecao.query('Codigo_UF!=0')
```

```python
# Espiando o conjunto de dados
projecao_UF
```

### **Scatterplot animado com indicadores das Projeções Demográficas 2010 a 2060**

```python
# Importando as bibliotecas necessarias
import plotly.express as px
```

```python
 # Construindo o Gráfico
px.scatter(
    data_frame = projecao_UF, 
    x="TMI", 
    y="EVN", 
    animation_frame="Ano", 
    animation_group="Nome_UF",
    size="Populacao_Total", 
    color="Nome_Regiao", 
    hover_name="Nome_UF",
    #log_x=True, 
    size_max=50, 
    range_x=[0,40], 
    range_y=[60,90]
)
```

### **Gráfico de barras animado da taxa de mortalidade infantil por grandes regiões do Brasil**

```python
# Filtrando somente as grandes regiões
projecao_regioes = projecao.query('(Codigo_UF==0) & (Codigo_Regiao!=0)')
```

```python
# Espiando o dataset
projecao_regioes
```

```python
#Fazendo gráfico de colunas animado da taxa de mortalidade infantil por Região
fig = px.bar(
    data_frame = projecao_regioes, 
    x="Nome_Regiao", 
    y="TMI", 
    color="Nome_Regiao",
    animation_frame="Ano", 
    animation_group="Nome_Regiao", 
    title = 'Taxa de Mortalidade Infantil por Região do Brasil, 2010-2060',
    labels={
      "Nome_Regiao": "Grande região do Brasil",
      "TMI": "Taxa de Mortalidade Infantil"
    },
    range_y=[0,30]
)

# Exibindo o gráfico
fig.show()
```
