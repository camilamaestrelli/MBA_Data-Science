## **Entendendo o contexto**

### **Títulos da Netflix**

Fundada no  EUA em 1997, inicialmente como uma empresa que entregava DVD pelo correios, a Netflix é uma grande empresa que oferece serviço de stream de filmes e programas de TV. Presente em dezenas de Países a Netflix já conta  em seu catálogo com mais de 8 mil títulos dos mais diversos gêneros.

### **O Objetivo da Aula**

### **O dataset**

Baseado em informações do catálogo de títulos da Netflix, o dataset, disponibilizado e  atualizado no Kaggle em outubro de 2021, este conjunto de dados foi estruturado de forma que cada instância representa um título da Netflix, e contém os seguintes atributos:

- **show_id:** Identificação única para a Produção;
- **type:** O tipo da Produção - Movie (filmes) ou TV Show (Programa de TV);
- **title:** Título da Produção;
- **director:** Diretor da Produção;
- **cast:**
- **country:** País de origen da Produção;
- **date_added:** Data de em que a Produção foi adicionada a Netflix;
- **release_year:** Ano de lançamento real da Produção;
- **rating:** Classificação;
- **duration:** Duração em minutos ou número de temporadas;
- **listed_in:** Gênero da Produção;
- **description:** Descrição resumida da Produção;

### **Gráfico de Pizza**

```python
# Importando as bibliotecas necessarias
import plotly.express as px
```

```python
 # Construindo o Gráfico de pizza com o tipo das produções
px.pie(data_frame = netflix, names = 'type')
```

```python
#Verificando se o plotly fez a distribuiçao correta
netflix['type'].value_counts(normalize=True).mul(100).round(1) 
```

```python
# Customizando as cores do gráficoss de pizza através de escalas prontas
px.pie(
    data_frame = netflix, 
    names = 'type',
    color = 'type',
    color_discrete_sequence=px.colors.sequential.Viridis
)   
```

```python
# Customizando as cores do gráficoss de pizza através de escalas customizada
px.pie(
    data_frame = netflix, 
    names = 'type',
    color = 'type',
    color_discrete_map = { 
      'Movie' : 'lightgray' , 
      'TV Show' : 'Orange'
    }
)                                 
```

```python
# Outras customizações de um gráfico de pizza
px.pie(
    data_frame = netflix, 
    names = 'type',
    color = 'type',
    color_discrete_map = { 
      'Movie' : 'green' , 
      'TV Show' : 'pink'
    },
    title = 'Distribuição da programação da Netflix por tipo',
    labels={'type':'Tipo da Produção'},
    opacity = 1,
    width=600, 
    height=400
)    
```

```python
 # Construindo o Gráfico de Rosca
px.pie(
    data_frame = netflix, 
    names = 'type',
    color = 'type',
    color_discrete_map = { 
      'Movie' : 'gray' , 
      'TV Show' : 'red'
    },
    title = '<b>Netflix</b><br>Distribuição da programação da Netflix por tipo',
    labels={'type':'Tipo da Produção'},
    opacity = 1,
    width=800, 
    height=600,
    hole=.5
)
```

Para mais detalhes e customização deste tipo de gráfico do Plotly consulte a documentação através do endereço: https://plotly.com/python/pie-charts/
