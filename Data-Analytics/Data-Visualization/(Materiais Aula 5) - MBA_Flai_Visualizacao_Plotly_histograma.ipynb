```python
# Importando o Pandas
import pandas as pd
```

```python
# Importando o conjunto de dados do ideb 2019 das escolas de ensino m√©dio
ideb = pd.read_csv(
    filepath_or_buffer = '/content/ideb_ensino_medio_escolas_2019.csv', 
    sep=';',
    decimal=',',
    encoding = 'utf-8',
    na_values = '-'
)
```

```python
# Espiando o conjunto de dados
ideb
```

### **Espiando a estrutura dos dados**

```python
# Informacoes gerais do conjunto de dados
ideb.info(verbose = True)
```

```python
# Verificando valores missing
ideb.isnull().sum()
```

### **Transforma√ß√µes no Conjunto de dados**

```python
# Retirando registros sem informa√ß√µes
ideb_medio = ideb[
          (ideb["nota_SAEB_2019_matematica"].notna()) & \
          (ideb["nota_SAEB_2019_lingua_portuguesa"].notna()) & \
          (ideb["rede"].notna())
]
```

```python
# Verificando valores missing
ideb_medio.isnull().sum()
```

```python
# Espiando o conjunto de dados
ideb_medio
```

### **Histogramas da Nota de Portugues e Matematica em 2019**

```python
# Importando as bibliotecas necessarias
import plotly.express as px
```

```python
 # Construindo histogramas
px.histogram(
    data_frame = ideb_medio, 
    x="nota_SAEB_2019_matematica", 
)
```

```python
 # Construindo histogramas
 # Limitando o n√∫mero de bins
px.histogram(
  data_frame = ideb_medio, 
  x="nota_SAEB_2019_matematica", 
  nbins=20
)
```

```python
 # Construindo histogramas
 # Mudando escala de contagem para probabilidade
px.histogram(
  data_frame = ideb_medio, 
  x="nota_SAEB_2019_matematica", 
  histnorm='probability'
)
```

```python
 # Construindo histogramas
 # Mudando escala de contagem para probabilidade
 # Estratificando por rede de ensino
px.histogram(
  data_frame = ideb_medio, 
  x="nota_SAEB_2019_matematica", 
  histnorm='probability',
  nbins = 50,
  color="rede"
)
```

```python
 # Construindo histogramas
 # Limitando o n√∫mero de bins
 # Mudando escala de contagem para probabilidade
 # Estratificando por rede de ensino
 # Inserindo gr√°ficos marginais com as distribui√ß√µes dos grupos de estratifica√ß√£o
px.histogram(
  data_frame = ideb_medio, 
  x="nota_SAEB_2019_matematica", 
  nbins=20,
  histnorm='probability',
  color="rede",
   marginal="box"
)
```

```python
 # Construindo histogramas
 # Limitando o n√∫mero de bins
 # Mudando escala de contagem para probabilidade
 # Estratificando por rede de ensino
 # Inserindo gr√°ficos marginais com as distribui√ß√µes dos grupos de estratifica√ß√£o
 # Separando a visualiza√ß√£o em colunas
px.histogram(
  data_frame = ideb_medio, 
  x="nota_SAEB_2019_matematica", 
  nbins=20,
  histnorm='probability',
  color="rede",
   marginal="box",
  facet_col="rede"
)
```

Para mais detalhes e customiza√ß√£o deste tipo de gr√°fico do Plotly consulte a documenta√ß√£o atrav√©s dos endere√ßos:

### **ECDF - Empirical Cumulative Distribution Function da Nota de Portugues e Matematica em 2019**

```python
 # Construindo ecdf da nota de matem√°tica
px.ecdf(
    data_frame = ideb_medio, 
    x="nota_SAEB_2019_matematica", 
)
```

```python
 # Construindo ecdf da nota de matem√°tica
px.ecdf(
    data_frame = ideb_medio, 
    x="nota_SAEB_2019_matematica", 
    color="rede"
)
```

```python
 # Construindo ecdf da nota de portugu√™s e matem√°tica
px.ecdf(
    data_frame = ideb_medio, 
   x=["nota_SAEB_2019_matematica", "nota_SAEB_2019_lingua_portuguesa"]
)
```

```python
 # Construindo ecdf da nota de portugu√™s e matem√°tica
 # Estratificando por rede de ensino
px.ecdf(
  data_frame = ideb_medio, 
  x=["nota_SAEB_2019_matematica", "nota_SAEB_2019_lingua_portuguesa"],
  facet_col="rede"
)
```
