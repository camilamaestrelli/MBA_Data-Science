## **Entendendo o contexto**

### **Índice de Desenvolvimento da Educação Básica 2019**

Já dizia William Edwards Deming: "O que não pode ser medido, não pode ser gerenciado." No Brasil, para avaliar gerenciar a qualidade do ensino 👨‍🏫 na educação básica, foi criado um indicador chamado de Índice de Desenvolvimento da Educação Básica. Criado em 2007, Ele reuni informações de dois importantes conceitos que é o fluxo escolar 👩‍🎓 e grau de conhecimento em portugês 📑 e matemática 📈 dos alunos da educação básica, que são aferidos nas provas SAEB - Sistema de Avaliação da Educação Básica.

```python
# Importando o Pandas
import pandas as pd
```

```python
# Importando o conjunto de dados do ideb 2019 das escolas de ensino médio
ideb = pd.read_csv(
    filepath_or_buffer = '/content/ideb_ensino_medio_escolas_2019.csv', 
    sep=';',
    decimal=',',
    encoding = 'utf-8',
    na_values = '-'
)
```

```python
# Espiando o conjunto de dados
ideb
```

### **Espiando a estrutura dos dados**

```python
# Informacoes gerais do conjunto de dados
ideb.info(verbose = True)
```

```python
# Verificando valores missing
ideb.isnull().sum()
```

### **Transformações no Conjunto de dados**

```python
# Retirando registros sem informações
ideb_medio = ideb[
          (ideb["nota_SAEB_2019_matematica"].notna()) & \
          (ideb["nota_SAEB_2019_lingua_portuguesa"].notna()) & \
          (ideb["rede"].notna())
]
```

```python
# Verificando valores missing
ideb_medio.isnull().sum()
```

```python
# Espiando o conjunto de dados
ideb_medio
```

### **Histogramas da Nota de Portugues e Matematica em 2019**

```python
# Importando as bibliotecas necessarias
import plotly.express as px
```

```python
 # Construindo histogramas
px.histogram(
    data_frame = ideb_medio, 
    x="nota_SAEB_2019_matematica", 
)
```

```python
 # Construindo histogramas
 # Limitando o número de bins
px.histogram(
  data_frame = ideb_medio, 
  x="nota_SAEB_2019_matematica", 
  nbins=20
)
```

```python
 # Construindo histogramas
 # Mudando escala de contagem para probabilidade
px.histogram(
  data_frame = ideb_medio, 
  x="nota_SAEB_2019_matematica", 
  histnorm='probability'
)
```

```python
 # Construindo histogramas
 # Mudando escala de contagem para probabilidade
 # Estratificando por rede de ensino
px.histogram(
  data_frame = ideb_medio, 
  x="nota_SAEB_2019_matematica", 
  histnorm='probability',
  nbins = 50,
  color="rede"
)
```

```python
 # Construindo histogramas
 # Limitando o número de bins
 # Mudando escala de contagem para probabilidade
 # Estratificando por rede de ensino
 # Inserindo gráficos marginais com as distribuições dos grupos de estratificação
px.histogram(
  data_frame = ideb_medio, 
  x="nota_SAEB_2019_matematica", 
  nbins=20,
  histnorm='probability',
  color="rede",
   marginal="box"
)
```

```python
 # Construindo histogramas
 # Limitando o número de bins
 # Mudando escala de contagem para probabilidade
 # Estratificando por rede de ensino
 # Inserindo gráficos marginais com as distribuições dos grupos de estratificação
 # Separando a visualização em colunas
px.histogram(
  data_frame = ideb_medio, 
  x="nota_SAEB_2019_matematica", 
  nbins=20,
  histnorm='probability',
  color="rede",
   marginal="box",
  facet_col="rede"
)
```

Para mais detalhes e customização deste tipo de gráfico do Plotly consulte a documentação através dos endereços: 
- https://plotly.com/python/histograms/

- https://plotly.com/python/reference/histogram/

### **ECDF - Empirical Cumulative Distribution Function da Nota de Portugues e Matematica em 2019**

```python
 # Construindo ecdf da nota de matemática
px.ecdf(
    data_frame = ideb_medio, 
    x="nota_SAEB_2019_matematica", 
)
```

```python
 # Construindo ecdf da nota de matemática
px.ecdf(
    data_frame = ideb_medio, 
    x="nota_SAEB_2019_matematica", 
    color="rede"
)
```

```python
 # Construindo ecdf da nota de português e matemática
px.ecdf(
    data_frame = ideb_medio, 
   x=["nota_SAEB_2019_matematica", "nota_SAEB_2019_lingua_portuguesa"]
)
```

```python
 # Construindo ecdf da nota de português e matemática
 # Estratificando por rede de ensino
px.ecdf(
  data_frame = ideb_medio, 
  x=["nota_SAEB_2019_matematica", "nota_SAEB_2019_lingua_portuguesa"],
  facet_col="rede"
)
```
