### **Transformações no Conjunto de dados**

```python
candidatos_para_merge = candidatos[['sequencial','nome','nome_urna','numero_partido',\
                                    'situacao','ocupacao','idade','genero','instrucao','estado_civil',\
                                    'nacionalidade','sigla_uf_nascimento','municipio_nascimento','raca']]                
```

```python
# juntado as bases de dados
eleicoes = pd.merge(
    left = resultados,
    right= candidatos_para_merge,
    how="left",
    left_on = 'sequencial_candidato',
    right_on = 'sequencial',
)
```

```python
# Informacoes gerais do conjunto de dados
eleicoes.info(verbose = True)
```

```python
# Espiando o conjunto de dados
eleicoes
```

### **Perfil dos Candidatos - eleições 2018**

```python
# Importando as bibliotecas necessarias
import matplotlib.pyplot as plt
```

#### **Distribuição dos candidatos por sexo**

```python
# Distribuição por sexo
perfil_sexo = eleicoes.genero.value_counts() \
  .rename_axis('sexo') \
  .to_frame('candidatos') \
  .reset_index()

# Espiando o dataset final
perfil_sexo
```

```python
# Fazendo um simples gráfico de pizza com a distribuição por sexo
plt.pie(x = perfil_sexo.candidatos, labels = perfil_sexo.sexo);
```

```python
# Inserindo as porcentagens no gráfico
plt.pie(
  x = perfil_sexo.candidatos,
  labels = perfil_sexo.sexo,
    autopct = '%.2f%%'
);
```

```python
# Criando gráfico de pizza mais customizado
plt.pie(
  x = perfil_sexo.candidatos,
  labels = perfil_sexo.sexo,
  autopct = '%.2f%%',
  shadow = True,
  colors = ['blue','magenta'],
  explode = [0,0.1],
  startangle = 90,
  textprops = dict(color ="white")
);
```

```python
# Customizando o rotulos dos valores

# Definindo uma função para customizar os rotulos
def autopct_custom(valor):
    def autopct_def(pct):
        total = sum(valor)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  (N={v:d})'.format(p=pct,v=val)
    return autopct_def

# Definindo o tamanho da figura
plt.figure(figsize=[10,7])

# Fazendo o gráfico de pizza
plt.pie(
  x = perfil_sexo.candidatos,
  labels = perfil_sexo.sexo,
  autopct = autopct_custom(perfil_sexo.candidatos),
  shadow = True,
  colors = ['lightgray','orange'],
  explode = [0,0.1],
  startangle = 90,
  textprops = dict(color ="black")
);

# Inserindo o título
plt.title('Sexo dos candidatos das eleições 2018');
```

```python
# Customizando o rotulos dos valores com função lambda

# Definindo o tamanho da figura
plt.figure(figsize=[10,7])

# Fazendo o gráfico de pizza
plt.pie(
  x = perfil_sexo.candidatos,
  labels = perfil_sexo.sexo,
  autopct = lambda p : '{:.2f}%  (N={:.0f})'.format(p,p * sum(perfil_sexo.candidatos)/100),
  shadow = True,
  colors = ['darkgreen','darkorange'],
  explode = [0,0.1],
  startangle = 90,
  textprops = dict(color ="white")
);

# Inserindo o título
plt.title('Sexo dos candidatos das eleições 2018');

```

#### **Distribuição dos candidatos por Estado Civil**

```python
# Distribuição dos candidatos por cargo
perfil_estcivil = eleicoes.estado_civil.value_counts() \
  .rename_axis('estado_civil') \
  .to_frame('candidatos') \
  .reset_index()

# Calculando porcentagem
perfil_estcivil['pct'] = (perfil_estcivil['candidatos'] / perfil_estcivil['candidatos'] .sum()) \
                    .mul(100).round(2)

# Espiando o dataset final
perfil_estcivil
```

```python
# Fazendo rápido um gráfico de Barras
plt.barh(perfil_estcivil.estado_civil  , perfil_estcivil.candidatos);
```

```python
# Customizando um gráfico de Barras

# Definindo o tamanho da figura
fig, ax = plt.subplots(figsize =(16, 9))

# Fazendo o gráfico de Colunas
ax.barh(
    y = perfil_estcivil.estado_civil , 
    width= perfil_estcivil.pct,
    height = 0.8,
    color = 'maroon',
    edgecolor="black"
);

# Inserindo o título
ax.set_title('Estado Civil dos candidatos das eleições 2018', loc ='left')

# Invertendo o eixo y
ax.invert_yaxis()

# Inserindo rótulos dos eixos
plt.xlabel("Proporção de Candidatos (em %)")
plt.ylabel("Estado Civil")

# Exibindo o gráfico
plt.show()
```

#### **Distribuição dos candidatos por Escolaridade**

```python
# Distribuição dos candidatos por cargo
perfil_instrucao = eleicoes.instrucao.value_counts() \
  .rename_axis('instrucao') \
  .reindex([
      'le e escreve', 
      'ensino fundamental incompleto', 
      'ensino fundamental completo', 
      'ensino medio incompleto',
      'ensino medio completo',
      'ensino superior incompleto',
      'ensino superior completo']) \
  .to_frame('candidatos') \
  .reset_index()

# Calculando porcentagem
perfil_instrucao['pct'] = (perfil_instrucao['candidatos'] / perfil_instrucao['candidatos'] .sum()) \
                    .mul(100).round(2)

# Espiando o dataset final
perfil_instrucao
```

```python
# Customizando um gráfico de Barras

# Definindo o tamanho da figura
fig, ax = plt.subplots(figsize =(16, 9))

# Fazendo o gráfico de Colunas
ax.bar(
    x = perfil_instrucao.instrucao  , 
    height= perfil_instrucao.candidatos,
    width = 0.8,
    color = 'orange',
    edgecolor="black"
);

# Inserindo o título
ax.set_title('Grau de Instrução dos candidatos das eleições 2018', loc ='left')

# Inserindo rótulos dos eixos
plt.xlabel("Grau de Instrução")
plt.ylabel("Número de Candidatos")

# Adicionando rótulo de valores
for i in ax.patches:
    plt.text(
        x = i.get_x()+0.3, 
        y = i.get_height()+50,
        s = str(round((i.get_height()), 2)),
       fontsize = 10, 
       fontweight ='bold',
       color ='black'
    )

# Rotacionando os rótulos do eixo x
plt.xticks(rotation=45, ha='right')

# Exibindo o gráfico
plt.show()
```

#### **Raça/cor dos Candidatos por Cargo**

```python
# Distribuição por sexo
perfil_raca = eleicoes.raca.value_counts() \
  .rename_axis('raca') \
  .reindex([
      'amarela', 
      'branca', 
      'indigena', 
      'parda',
      'preta']) \
  .to_frame('candidatos') \
  .reset_index()

# Calculando porcentagem
perfil_raca['pct'] = (perfil_raca['candidatos'] / perfil_raca['candidatos'] .sum()) \
                    .mul(100).round(2)

# Espiando o dataset final
perfil_raca
```

```python
# Customizando um gráfico de Colunas

# Definindo o tamanho da figura
fig, ax = plt.subplots(figsize =(16, 9))

# Fazendo o gráfico de Colunas
ax.bar(
    x = perfil_raca.raca  , 
    height= perfil_raca.pct,
    width = 0.8,
    color = 'purple',
    edgecolor="black"
);

# Inserindo o título
ax.set_title('Raça / Cor dos candidatos das eleições 2018', loc ='left')

# Inserindo rótulos dos eixos
plt.xlabel("Raça / Cor")
plt.ylabel("Proporção de Candidatos (%)")

# Adicionando rótulo de valores
for i in ax.patches:
    plt.text(
        x = i.get_x()+0.3, 
        y = i.get_height()+1,
        s = str(round((i.get_height()), 2)),
       fontsize = 10, 
       fontweight ='bold',
       color ='black'
    )

# Exibindo o gráfico
plt.show()
```
