# **Campeonato Brasileiro de Futebol**

## **Entendendo o contexto**

### **Brasileirao**

Apesar do futebol ter nascido na Inglaterra, o Brasil é considerado o país do Futebol, temos muitos amantes deste esporte entre o povo brasileiro. Como esperado em uma nação com povo tão apaixonado pela bola, temos por aqui o Campeonato Brasileiro de Futebol, também conhecido como Brasileirão, é disputado por 128 times, separados em quatro divisões: Série A (20 times), Série B (20 times), Série C (20 times) e Série D (68 times).

Anualmente ocorre uma edição deste campeonato. Ao final, os quatro últimos colocados da Série A passam a compor a Série B e os quatro primeiros colocados da Série B passam a compor Série A. Também, os quatro últimos colocados da Série B passam a compor a Série C e os quatro primeiros colocados da Série C passam a compor Série D. A Série D tem ingresso baseado nos campeonatos estaduais.

Desde 2003 a Série A do Brasileirão é disputado no sistema de pontos corridos com turno e returno, ou seja, cada time enfrenta os demais duas vezes na edição do campeonato. Se ganhar o time soma 3 pontos, se empatar soma 1 ponto e se perder não soma ponto na rodada. São 38 rodadas no campeonato. A classificação é determinada em ordem decrescente pelo total de pontos somados por cada time ao longo das rodadas. Caso haja empate na soma de pontos, os critérios de desempate são:
1. Maior número de vitórias.
1. Maior saldo de gols.
1. Maior número de gols a favor.
1. Confronto direto.
1. Menor número de cartões vermelhos recebidos.
1. Menor número de cartões amarelos recebidos.
1. Sorteio.

Vale ressaltar que 24 times disputaram a Série A do Brasileirão de 2003 e 2004. Já em 2005 foram 22 times que disputando este campeonato. E, à partir de 2006, 20 times passaram a disputar a "primeira divisão" do Campeonato Brasileiro de Futebol do Brasil.

### **Transformação do conjunto de dados**

```python
# Criando a variável total de gols da partida
cbf['gols_partida'] = cbf['gols_man'] + cbf['gols_vis']
```

```python
# Espiando o dataset com a nova variável
cbf[['gols_man','gols_vis','gols_partida']].head(15)
```

```python
# Definindo função para informar resultado da partida
def funcao_resultado(x):
    if x == 0:
        return 'Empate'
    elif x > 0:
        return 'Vitória do Mandante'
    elif x < 0:
        return 'Vitória do Visitante'
```

```python
# Criando o atributo com o resultado da partida
cbf['resultado_da_partida'] = (cbf['gols_man'] - cbf['gols_vis']).apply(funcao_resultado)
```

```python
# Espiando o dataset com a nova variável
cbf[['gols_man','gols_vis','resultado_da_partida']].head(15)
```

### **Análise da Série A do Brasileirão**

```python
# Importando as bibliotecas necessarias
import altair as alt
```

```python
#### **Resultados das partidas**
cbf.shape
```

#### **Resultados das partidas** 

```python
# Resultados da partidas por ano
resultados_partidas = cbf[['resultado_da_partida']] \
  .value_counts(normalize=True,)\
  .mul(100)\
  .round(1)\
  .reset_index()\
  .rename(columns={0: "partidas"})


# Espiando o resultado
resultados_partidas
```

```python

alt.Chart(data=resultados_partidas).mark_arc().encode(
    theta=alt.Theta(field="partidas", type="quantitative"),
    color=alt.Color(field="resultado_da_partida", type="nominal"),
)
```

```python
# Definindo parâmetros básicos do gráfico
base = alt.Chart(
            data = resultados_partidas,
            title = 'Resultados das partilas do Brasileirão, 2003-2020.'
          )\
          .encode(
            theta=alt.Theta(shorthand = "partidas:Q", stack=True), 
            color=alt.Color(
                shorthand ="resultado_da_partida:N", 
                legend = alt.Legend(title = 'Resultado da Partida (em %)'),
                #scale=alt.Scale(scheme='set1')
                scale=alt.Scale(range=["orange", "darkgreen", "red"])
            )
          )

# Criando a pizza
pizza = base.mark_arc(outerRadius=120)

# Criando os Labels
texto = base.mark_text(radius=140, size=15).encode(text="partidas:Q")

# Exibindo o gráfico
pizza+texto
```

#### **Resultados das partidas por ano** 

```python
# Resultados da partidas por ano
resultados_partidas_por_ano = cbf[['ano_campeonato','resultado_da_partida']] \
  .groupby(['ano_campeonato','resultado_da_partida'], as_index = False)\
  .size() \
  .rename(columns={"size": "partidas"})

# Espiando o resultado
resultados_partidas_por_ano.head()
```

```python
# Fazendo gráfico de colunas empilhadas
alt.Chart(data = resultados_partidas_por_ano)\
  .mark_bar(width=20)\
  .encode(
    x=alt.X(shorthand = 'ano_campeonato',title = 'Ano do Campeonato'),
    y=alt.Y(shorthand = 'partidas', stack="normalize", title = 'Quantidade de Partidas'),
    color='resultado_da_partida',
    tooltip=['ano_campeonato', 'resultado_da_partida', 'partidas']
).properties(width=600, height=400).interactive()
```

#### **Idade Media da equipe titular mandante x Valor da equipe titular mandante**

```python
# Campeonato em 2018
cbf_2018 = cbf.query('ano_campeonato==2018')
```

```python
# Elaborando o diagrama de dispersão
alt.Chart(data = cbf_2018) \
  .mark_circle(size=60)\
  .encode(
    x=alt.X(
        shorthand = 'idade_media_titular_man:Q',
        scale=alt.Scale(domain=(23, 35)),
        title = 'Idade média dos titulares do time mandante'
    ),
    y=alt.Y(
        shorthand = 'valor_equipe_titular_man:Q',
        title = 'Valor da equipe titular do time mandante'
    ),
    color=alt.Color(
      shorthand ="time_man:N", 
      legend = alt.Legend(title = 'Time mandante'),
      scale=alt.Scale(scheme='tableau20')
    ),
    tooltip=['time_man', 'idade_media_titular_man', 'valor_equipe_titular_man']
).properties(width=600, height=400).interactive()
```

#### **Gols por ano**

```python
# Gols por ano
gols_por_ano = cbf[['ano_campeonato','gols_man','gols_vis','gols_partida']].query('ano_campeonato>2005') \
  .groupby(['ano_campeonato'], as_index = True)\
  .sum()\
  .reset_index()

# Espiando o resultado
gols_por_ano.head()
```

```python
# Fazendo gráfico de gols por partida
alt.Chart(data=gols_por_ano).mark_area(
    line={'color':'darkgreen'},
    color=alt.Gradient(
        gradient='linear',
        stops=[alt.GradientStop(color='yellow', offset=0),
               alt.GradientStop(color='white', offset=1)],
        x1=1,
        x2=0,
        y1=1,
        y2=0       
    )
).encode(
    alt.X('ano_campeonato:Q'),
    alt.Y('gols_partida:Q')
).properties(width=600, height=400)
```

#### **Gols por ano e mando de campo**

```python
# Colocando o 'ano_campeonato' como índice do dataframe
gols_por_tipo_ano = gols_por_ano.set_index('ano_campeonato')
```

```python
# Colocando no formato longer
gols_por_tipo_ano = gols_por_tipo_ano.stack()
gols_por_tipo_ano.head(6)
```

```python
# Transformando em dataframe
gols_por_tipo_ano = pd.DataFrame(gols_por_tipo_ano)

# Renomeando o índice
gols_por_tipo_ano.index.names = ['ano_campeonato','tipo_gols']

# Renomeando o número de gols que está como variável '0'
gols_por_tipo_ano = gols_por_tipo_ano.rename(columns={0: "gols"})

# Transformando os índices em colunas do dataframe
gols_por_tipo_ano = gols_por_tipo_ano.reset_index()

# Espiando o resultado
gols_por_tipo_ano.head()
```

```python
# Fazendo o gráfico de número de gols por mando de campo
alt.Chart(
    data = gols_por_tipo_ano,
    title = 'Quantidade de gols por edição do Brasileirão, por mando de campo.',
    height = 400,
    width = 400
  )\
  .mark_area(opacity = 0.3)\
  .encode(
    x=alt.X(shorthand = "ano_campeonato:Q", title = 'Ano do Campeonato'),
    y=alt.Y("gols:Q", stack=None, title = 'Total de Gols na edição do Brasileirão'),
    color="tipo_gols:N"
  )
```

#### **Campeões do Brasileirão na década 2011-2020**

```python
partida_do_campeao = cbf.query('(ano_campeonato>2010) & (rodada==38) & ((colocacao_man==1) | (colocacao_vis==1))')
```

```python
# Definindo função para pegar o campeao
def funcao_campeao(a,b,c):
    if a < 0:
        return b
    else:
        return c
```

```python
partida_do_campeao['campeao'] = partida_do_campeao\
  .apply(lambda x: funcao_campeao(x['colocacao_man'] - x['colocacao_vis'],x['time_man'],x['time_vis']),axis=1);
```

```python
partida_do_campeao[['ano_campeonato','time_man','time_vis','colocacao_man','colocacao_vis','campeao']]
```

```python
# Campeos das ultimas 10 edições do Brasileirão registrada neste dataset
campeoes_decada_2010 = partida_do_campeao[['campeao']] \
  .value_counts()\
  .reset_index()\
  .rename(columns={0: "campeonatos"})


# Espiando o resultado
campeoes_decada_2010
```

```python
# Definindo as barras
alt.Chart(
    data=campeoes_decada_2010,
    width = 600,
    height = 400, 
    title = 'Quantidade de edições vitoriosas dos campões do Brasileirão na década 2011-2020.'
  ).mark_bar(
      opacity=0.8,
      color='darkgreen'
  ).encode(
    x=alt.X(
        shorthand = "campeao:N", 
        sort='-y', 
        axis = alt.Axis(title = 'Time Campeão')
    ),
    y=alt.Y(
        shorthand = "campeonatos:Q", 
        axis = alt.Axis(title = 'Quantidade de Edições'),
        scale=alt.Scale(domain=(0,4)),
    )    
)
```
