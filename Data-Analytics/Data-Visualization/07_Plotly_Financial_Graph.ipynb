# Gráfico Financeiro com Plotly

## Ações - PETR4

### Webscraping com pandas_datareader


```python
!pip install --upgrade pandas
!pip install --upgrade pandas-datareader

from pandas_datareader import data as web
import numpy as np
import pandas as pd
from datetime import date   
import plotly.graph_objects as go

```

```python
cotacao_petr4 = web.DataReader('PETR4.SA', data_source='yahoo', 
                               start = '01-01-1998', 
                               end = date.today() # pega os dados até hoje
                               )
cotacao_petr4
```

```python
cotacao_petr4.info(verbose = True)
```

* Nota-se que não existem campos vazios

### Criando Gráfico estilo ações - Candlestick

```python
grafico_petr4 = go.Figure(
    data = [go.Candlestick(
        x = cotacao_petr4.index,
        open = cotacao_petr4.Open,
        close = cotacao_petr4.Close,
        high = cotacao_petr4.High,
        low = cotacao_petr4.Low, 
        name = 'PETR4'
    )
    ]
)

grafico_petr4.update_layout(xaxis_rangeslider_visible = False)

grafico_petr4.show()
```

### Analisando Média Móvel

* Percebe-se uma grande variação no dia 5 de agosto de 2003

* Observa-se um pico entre 4 e 5 de agosto de 2002.
* Isso se deu provavelmente a um split da ação, onde a quantidade de ações em circulação de uma empresa é aumentada, por meio da divisão do número original de ações, reduzindo seu valor nominal.
* E o valor 'splitado' não foi atualizado

```python
cotacao_petr4['MM20'] = cotacao_petr4['Close'].rolling(window=20).mean() # média móvel de 20 dias
cotacao_petr4['MM5'] = cotacao_petr4['Close'].rolling(window=5).mean() # média móvel de 5 dias
cotacao_petr4
```

```python
cotacao_petr4.loc[cotacao_petr4.index == '2003-08-05']
```

```python
cotacao_petr4.loc[cotacao_petr4.index == '2003-08-04']
```

```python
grafico_petr4.add_trace(go.Scatter(x=cotacao_petr4.index, 
                         y=cotacao_petr4['MM5'], 
                         opacity=0.7, 
                         line=dict(color='blue', width=2), 
                         name='MM 5 PETR4'))
grafico_petr4.add_trace(go.Scatter(x=cotacao_petr4.index, 
                         y=cotacao_petr4['MM20'], 
                         opacity=0.7, 
                         line=dict(color='orange', width=2), 
                         name='MM 20 PETR4'))
grafico_petr4.update_layout(xaxis_rangeslider_visible = False, template='simple_white')
```

### Adicionando volume ao grafico

```python
from plotly.subplots import make_subplots

grafico_petr4_volume = make_subplots(rows=2, cols=1, shared_xaxes=True, 
               vertical_spacing=0.03, subplot_titles=('PETR4', 'Volume'), 
               row_width=[0.2, 0.7])

# Plot OHLC on 1st row
grafico_petr4_volume.add_trace(go.Candlestick(x=cotacao_petr4.index, open=cotacao_petr4.Open, high=cotacao_petr4.High,
                low=cotacao_petr4.Low, close=cotacao_petr4.Close, name="PETR4"), 
                row=1, col=1
)
grafico_petr4_volume.add_trace(go.Scatter(x=cotacao_petr4.index, 
                         y=cotacao_petr4['MM5'], 
                         opacity=0.7, 
                         line=dict(color='blue', width=2), 
                         name='MM 5 PETR4'),
                         row=1, col=1)
grafico_petr4_volume.add_trace(go.Scatter(x=cotacao_petr4.index, 
                         y=cotacao_petr4['MM20'], 
                         opacity=0.7, 
                         line=dict(color='orange', width=2), 
                         name='MM 20 PETR4'),
                         row=1, col=1)

# Bar trace for volumes on 2nd row without legend
grafico_petr4_volume.add_trace(go.Bar(x=cotacao_petr4.index, y=cotacao_petr4['Volume'],
                               showlegend=False), row=2, col=1)

# Do not show OHLC's rangeslider plot 
grafico_petr4_volume.update(layout_xaxis_rangeslider_visible=False)
grafico_petr4_volume.update_layout(template='plotly')
grafico_petr4_volume.show()
```

* Nota-se uma diminuição no volume de transações das ações PETR4 no mercado

## BITCOIN VS ETHEREUM

```python
bitcoin = web.DataReader('BTC-USD', data_source='yahoo', 
                               start = '01-01-1998', 
                               end = date.today())
bitcoin
```

```python
ethereum = web.DataReader('ETH-USD', data_source='yahoo', 
                               start = '01-01-1998', 
                               end = date.today())
ethereum
```

```python
grafico_bitcoin = go.Figure(
    data = [go.Candlestick(
        x = bitcoin.index,
        open = bitcoin.Open,
        close = bitcoin.Close,
        high = bitcoin.High,
        low = bitcoin.Low,
        name = 'bitcoin'
    )
    ]
)

grafico_bitcoin.add_trace(go.Candlestick(
        x = ethereum.index,
        open = ethereum.Open,
        close = ethereum.Close,
        high = ethereum.High,
        low = ethereum.Low,
        increasing_line_color= 'cyan', decreasing_line_color= 'gray',
        name='ethereum'
    )
)

grafico_bitcoin.update_layout(xaxis_rangeslider_visible = False, template='plotly_dark')

grafico_bitcoin.show()
```

* Observa-se uma grande diferença nas 2 criptomoedas, não sendo possível sua comparação. Assim, vamos tranformar o eixo y em uma escla logaritmica


* A escala logarítmica irá achatar o gráfico. Dessa forma, teremos uma comparação de comportamentos

* Com isso, teremos um melhor entendimento da variabilidade em detrimento do valor absoluto

```python
grafico_bitcoin_log = go.Figure(
    data = [go.Candlestick(
        x = bitcoin.index,
        open = np.log(bitcoin.Open),
        close = np.log(bitcoin.Close),
        high = np.log(bitcoin.High),
        low = np.log(bitcoin.Low),
        name = 'bitcoin'
    )
    ]
)

grafico_bitcoin_log.add_trace(go.Candlestick(
        x = ethereum.index,
        open = np.log(ethereum.Open),
        close = np.log(ethereum.Close),
        high = np.log(ethereum.High),
        low = np.log(ethereum.Low),
        increasing_line_color= 'cyan', decreasing_line_color= 'gray',
        name='ethereum'
    )
)

grafico_bitcoin_log.update_layout(xaxis_rangeslider_visible = False, template='plotly_dark')

grafico_bitcoin_log.show()
```

* quando analisamos a variabilidade das 2 criptos, percebemos uma semelhança nas 2 séries temporais
