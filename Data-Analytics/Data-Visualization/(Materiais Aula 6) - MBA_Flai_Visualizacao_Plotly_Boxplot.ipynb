### **Gr√°fico dos Sal√°rios**

```python
# Importando as bibliotecas necessarias
import plotly.express as px
```

```python
# Selecionando somente os 95% dos trabalhadores com menores sal√°rios
# Retirando os 5% dos trabalhadores com maiores sal√°rios
rh_eda = rh.query('contrato_salario<4593.520500')
```

```python
# Verificando a distribui√ß√£o dos sal√°rios pelo boxplot
px.box(data_frame = rh_eda, y="contrato_salario", width=800, height = 600)
```

```python
# Estratificando a an√°lise dos sal√°rios pelo sexo dos trabalhadores
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    x = 'func_sexo',
    template = 'ggplot2',
    width=800, 
    height = 600
)
```

```python
# Estratificando a an√°lise dos sal√°rios pelo Ra√ßa/Cor dos trabalhadores
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'func_racacor',     
    title = 'Distribui√ß√£o dos sal√°rios por Ra√ßa/Cor do trabalhador',
    template = 'plotly_white',
    width=800, 
    height = 600
)
```

```python
# Estratificando a an√°lise dos sal√°rios Por exist√™ncia de Defici√™ncia do trabalhadores
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'func_deficiencia',  
    color_discrete_map = { 
      'NAO' : 'red',
      'SIM' : 'green'
    },   
    title = 'Distribui√ß√£o dos sal√°rios por exist√™ncia de Defici√™ncia do trabalhadores',
    labels = {
      'contrato_salario': 'Sal√°rio (em reais)',
      'func_deficiencia': 'Trabalhador possui alguma Defici√™ncia'
    },
    template = 'none',
    width=800, 
    height = 600
)
```

```python
# Estratificando a an√°lise dos sal√°rios pela escolaridade dos trabalhadores
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'func_escolaridade',     
    title = 'Distribui√ß√£o dos sal√°rios por escolaridade do trabalhador',
    labels = {
      'contrato_salario': 'Sal√°rio (em reais)',
      'func_escolaridade': 'Escolaridade'
    },
    template = 'plotly_dark',
    category_orders = {'func_escolaridade':['ANALFABETO-FUND_INCOMPLETO','FUND_COMPLETO-MEDIO_INCOMPLETO','MEDIO COMPLETO','SUPERIOR_INCOMPLETO','SUPERIOR_COMPLETO']},
    width=800, 
    height = 600
)
```

```python
# Estratificando a an√°lise dos sal√°rios pela Unidade da Federa√ß√£o dos trabalhadores
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'func_uf',     
    title = 'Distribui√ß√£o dos sal√°rios por UF de Resid√™ncia do Trabalhador',
    labels = {
      'contrato_salario': 'Sal√°rio (em reais)',
      'func_uf': 'UF'
    },
    template = 'presentation',
    width=800, 
    height = 600
)
```

```python
# Estratificando a an√°lise dos sal√°rios pelo setor da economia da Empresa
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'empresa_setor',     
    title = 'Distribui√ß√£o dos sal√°rios por Setor da Economia da Empresa',
    labels = {
      'contrato_salario': 'Sal√°rio (em reais)',
      'empresa_setor': 'Setor'
    },
    template = 'xgridoff',
    width=800, 
    height = 600
)
```

```python
# Estratificando a an√°lise dos sal√°rios pelo tempo que pediu demiss√£o
px.box(rh_eda, y="contrato_salario", color = 'turnover_apos_1_ano')
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'turnover_apos_1_ano',     
    title = 'Distribui√ß√£o dos sal√°rios por tempo at√© o turnover',
    labels = {
      'contrato_salario': 'Sal√°rio (em reais)',
      'turnover_apos_1_ano': 'Turnover ap√≥s 1 ano'
    },
    template = 'gridon',
    width=800, 
    height = 600
)
```
