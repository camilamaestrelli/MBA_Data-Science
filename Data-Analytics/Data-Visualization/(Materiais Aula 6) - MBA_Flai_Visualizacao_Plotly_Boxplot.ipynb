# **Análise da distribuição Salarial**

## **Entendendo o contexto**

### **Salários**

Ao desenvolver funções e tarefas para determinada organização, através de algum tipo de contrato o funcionário faz um acordo com a empresa para definir qual será sua remuneração 💲, que pode ser composta por um salário fixo, bem como haver outras fontes nesta remuneração como adicional noturno, pagamento de horas extras, adicional de periculosidade; participação nos lucros (PL), participação acionária, insalubridade, gorjetas e gratificações, gastos de viagens e hospedagens a serviço, comissões e percentagens, incentivos da corporação, entre outros.

### **O Objetivo da Aula**

Entender o comportamento dos salários através de visualização de dados.

### **O dataset**

Este conjunto de dados foi estruturado de forma que cada instância é um dia de determinado ano, e os atributos são indicadores de preços de Criptmoedas:

- **func_sexo:** Sexo do trabalhador;

- **func_idade:** Idade do trabalhador;

- **func_racacor:** Raça/Cor do trabalhador;

- **func_escolaridade:** Escolaridade do trabalhador;

- **func_uf:** UF de residência do trabalhador;

- **func_deficiencia:** Se o trabalhador tem alguma deficiência;

- **empresa_porte:** Porte da Empresa;

- **empresa_setor:** Setor de Atividade Econômica da Empresa;

- **contrato_horastrabalho:** Horas do contrato de trabalho;

- **contrato_salario:** Salário (em reais);

- **turnover_apos_1_ano:** Se no momento da demissão tinha mais de um ano Empresa;

O conjunto de dados está disponível em um repositório no GitHub que pode ser acessado através do link:

- https://raw.githubusercontent.com/jonates/opendata/master/turnover/turnover_treino.csv

### **Gráfico dos Salários**

```python
# Importando as bibliotecas necessarias
import plotly.express as px
```

```python
# Selecionando somente os 95% dos trabalhadores com menores salários
# Retirando os 5% dos trabalhadores com maiores salários
rh_eda = rh.query('contrato_salario<4593.520500')
```

```python
# Verificando a distribuição dos salários pelo boxplot
px.box(data_frame = rh_eda, y="contrato_salario", width=800, height = 600)
```

```python
# Estratificando a análise dos salários pelo sexo dos trabalhadores
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    x = 'func_sexo',
    template = 'ggplot2',
    width=800, 
    height = 600
)
```

```python
# Estratificando a análise dos salários pelo Raça/Cor dos trabalhadores
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'func_racacor',     
    title = 'Distribuição dos salários por Raça/Cor do trabalhador',
    template = 'plotly_white',
    width=800, 
    height = 600
)
```

```python
# Estratificando a análise dos salários Por existência de Deficiência do trabalhadores
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'func_deficiencia',  
    color_discrete_map = { 
      'NAO' : 'red',
      'SIM' : 'green'
    },   
    title = 'Distribuição dos salários por existência de Deficiência do trabalhadores',
    labels = {
      'contrato_salario': 'Salário (em reais)',
      'func_deficiencia': 'Trabalhador possui alguma Deficiência'
    },
    template = 'none',
    width=800, 
    height = 600
)
```

```python
# Estratificando a análise dos salários pela escolaridade dos trabalhadores
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'func_escolaridade',     
    title = 'Distribuição dos salários por escolaridade do trabalhador',
    labels = {
      'contrato_salario': 'Salário (em reais)',
      'func_escolaridade': 'Escolaridade'
    },
    template = 'plotly_dark',
    category_orders = {'func_escolaridade':['ANALFABETO-FUND_INCOMPLETO','FUND_COMPLETO-MEDIO_INCOMPLETO','MEDIO COMPLETO','SUPERIOR_INCOMPLETO','SUPERIOR_COMPLETO']},
    width=800, 
    height = 600
)
```

```python
# Estratificando a análise dos salários pela Unidade da Federação dos trabalhadores
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'func_uf',     
    title = 'Distribuição dos salários por UF de Residência do Trabalhador',
    labels = {
      'contrato_salario': 'Salário (em reais)',
      'func_uf': 'UF'
    },
    template = 'presentation',
    width=800, 
    height = 600
)
```

```python
# Estratificando a análise dos salários pelo setor da economia da Empresa
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'empresa_setor',     
    title = 'Distribuição dos salários por Setor da Economia da Empresa',
    labels = {
      'contrato_salario': 'Salário (em reais)',
      'empresa_setor': 'Setor'
    },
    template = 'xgridoff',
    width=800, 
    height = 600
)
```

```python
# Estratificando a análise dos salários pelo tempo que pediu demissão
px.box(rh_eda, y="contrato_salario", color = 'turnover_apos_1_ano')
px.box(
    data_frame = rh_eda, 
    y="contrato_salario", 
    color = 'turnover_apos_1_ano',     
    title = 'Distribuição dos salários por tempo até o turnover',
    labels = {
      'contrato_salario': 'Salário (em reais)',
      'turnover_apos_1_ano': 'Turnover após 1 ano'
    },
    template = 'gridon',
    width=800, 
    height = 600
)
```
