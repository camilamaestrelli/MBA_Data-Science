## 1) Definição do índice e 'limpeza de variáveis'

```python
df = df.set_index('ID')
df.loc[:, 'Idade'] = 2020 - df.Year_Birth
df.loc[:, 'Primeira_compra'] = (pd.to_datetime('2020-12-15') - pd.to_datetime(df.Dt_Customer)).dt.days
df = df.drop(['Year_Birth','Dt_Customer'],axis=1)
df
```

## 2) Separação em variáveis preditivas e variável resposta

```python
X = df.drop('Response',axis=1)
y = df.Response
```

## 3) Separação em treino e teste
 - Neste caso, validação <i>out-of-sample</i>

```python
from sklearn.model_selection import train_test_split

X_tr, X_ts, y_tr, y_ts = train_test_split(X,y, test_size=0.30)
```

## 4) Conversão de tipo das variáveis categóricas

```python
possiveis_Education = pd.CategoricalDtype(['Basic','2n Cycle','Graduation','Master','PhD'],ordered=True)
possiveis_Marital_Status = pd.CategoricalDtype(X_tr.Marital_Status.unique())

possiveis_Education
```

```python
possiveis_Marital_Status
```

```python
X_tr.loc[:,'Education'] = X_tr.Education.astype(possiveis_Education)
X_ts.loc[:,'Education'] = X_ts.Education.astype(possiveis_Education)

X_tr.loc[:,'Marital_Status'] = X_tr.Marital_Status.astype(possiveis_Marital_Status)
X_ts.loc[:,'Marital_Status'] = X_ts.Marital_Status.astype(possiveis_Marital_Status)

X_tr
```

## 5) One-hot encoding de variáveis categóricas

```python
X_tr_d = pd.get_dummies(X_tr,columns=['Education','Marital_Status'],drop_first=True)
X_ts_d = pd.get_dummies(X_ts,columns=['Education','Marital_Status'],drop_first=True)
X_tr_d
```

## 6) Verificação de valores faltantes

```python
X_tr_d.isnull().sum()
```

```python
X_ts_d.isnull().sum()
```

## 7) Imputação de valores faltantes

 - Note que sempre <u>aprendemos no treino</u> e <u>aplicamos no teste</u>

```python
valor_para_imputar = X_tr_d.Income.median()
X_tr_d.loc[X_tr_d.Income.isnull(),'Income'] = valor_para_imputar
X_ts_d.loc[X_ts_d.Income.isnull(),'Income'] = valor_para_imputar
```

```python
from sklearn.preprocessing import StandardScaler

numericas = [
    'Income',
    'Recency',
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds',
    'NumDealsPurchases',
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases',
    'NumWebVisitsMonth',
    'Idade',
    'Primeira_compra',
]

scalers = {}
X_tr_s = X_tr_d.copy()
X_ts_s = X_ts_d.copy()
for col in numericas:
    scaler = StandardScaler()
    scaler.fit(X_tr_s[[col]])
    X_tr_s.loc[:,col] = scaler.transform(X_tr_s[[col]])
    X_ts_s.loc[:,col] = scaler.transform(X_ts_s[[col]])
    scalers[col] = scaler
    
X_tr_s
```

```python
from sklearn.preprocessing import KBinsDiscretizer

scalers_k = {}
X_tr_k = X_tr_d.copy()
X_ts_k = X_ts_d.copy()
for col in numericas:
    scaler = KBinsDiscretizer(n_bins=5, encode='onehot-dense', strategy='quantile')
    scaler.fit(X_tr_k[[col]])
    
    novas_colunas = []
    for comeco in scaler.bin_edges_[0][:-1]:
        novas_colunas.append(f'{col}_{comeco}')
    novas_colunas
    
    auxiliar = pd.DataFrame(scaler.transform(X_tr_k[[col]]),columns=novas_colunas, index=X_tr_k.index)
    X_tr_k = X_tr_k.join(auxiliar)
    auxiliar = pd.DataFrame(scaler.transform(X_ts_k[[col]]),columns=novas_colunas, index=X_ts_k.index)
    X_ts_k = X_ts_k.join(auxiliar)
    
    X_tr_k = X_tr_k.drop(col,axis=1)
    X_ts_k = X_ts_k.drop(col,axis=1)
    
    scalers_k[col] = scaler
    
X_tr_k
```

# Correlação

```python
corr_cat = X_tr_f2.corr('spearman')
corr_cat
```

```python
from matplotlib.colors import TwoSlopeNorm

plt.figure(figsize=(15,15))
plt.imshow(corr_cat,cmap='bwr',norm=TwoSlopeNorm(0.7,vmin=0,vmax=1))
plt.colorbar()
```

```python
from sklearn.feature_selection import mutual_info_classif

X_tr_f3 = X_tr_f2.copy()
X_ts_f3 = X_ts_f2.copy()
dropped = []

for num1,col1 in enumerate(corr_cat.columns):
    if col1 in dropped:
        continue
    for num2,col2 in enumerate(corr_cat.index):
        if col2 in dropped:
            continue
        if num2 <= num1:
            continue
        if col1==col2:
            continue
        val = corr_cat.loc[col2,col1]
        if val < .7:
            continue
            
        imm1 = mutual_info_classif(X_tr_f3[[col1]],y_tr)/mutual_info_classif(y_tr.values.reshape(-1,1),y_tr)
        imm2 = mutual_info_classif(X_tr_f3[[col2]],y_tr)/mutual_info_classif(y_tr.values.reshape(-1,1),y_tr)
        
        if imm1 > imm2:
            dropped.append(col2.strip())
            print(f'Entre {col1} e {col2}, descartamos {col2}')
        else:
            dropped.append(col1.strip())
            print(f'Entre {col1} e {col2}, descartamos {col1}')

X_tr_f3 = X_tr_f3.drop(dropped,axis=1)
X_ts_f3 = X_ts_f3.drop(dropped,axis=1)
X_tr_f3
```

# Regularização L1

# CUIDADO

Para usar a regularização L1, você tem que alterar o parâmetro `solver` da LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# C = 1/alpha
params = {
    'C':[0.1,1,10,50]
}

lr2 = LogisticRegression(penalty='l1', solver='saga', max_iter=10_000)
grid = GridSearchCV(lr2, params, cv=10, scoring='roc_auc', n_jobs=-1, verbose=10)
grid.fit(X_tr_f3, y_tr)
```

```python
grid.best_params_
```

```python
roc_auc_score(y_ts, grid.predict_proba(X_ts_f3)[:,1])
```
